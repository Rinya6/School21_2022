У нас большая выборками (n≥30), мы можем предположить, что выборочное распределение выборочного среднего нормально распределено (благодаря центральной предельной теореме), и вместо этого можем использовать функцию norm.interval() из scipy библиотека .stats.

```python
import scipy.stats as st
```

```python
def trust_interval(sample, alpha):
  trust_interval = st.norm.interval(alpha=alpha, loc= np.mean(sample), scale=st.sem(sample))
  return print('Доверительный интервал {}%'.format(int(alpha*100)),'= [{} - {}]'.format(trust_interval[0].round(5), trust_interval[1].round(5)))
```

```python
# Доверительный интервал 95%, для выборки в 100 человек

sample_100 = np.random.choice(all_people, size=100)
ti_95_100 = trust_interval(sample_100, 0.95)
ti_95_100
```

```python
# Доверительный интервал 95%, для выборки в 1000 человек

sample_1000 = np.random.choice(all_people, size=1000)
ti_95_1000 = trust_interval(sample_1000, 0.95)
ti_95_1000
```

```python
# Доверительный интервал 95%, для выборки в 10000 человек

sample_10000 = np.random.choice(all_people, size=10000)
ti_95_10000 = trust_interval(sample_10000, 0.95)
ti_95_10000
```

Формула стандартного нормального распределения (распределения Гаусса):

- [Нормальное распределение, ВикипедиЯ](https://ru.wikipedia.org/wiki/Нормальное_распределение)

Известно, что в конкурирующем банке у клиентов средний остаток на текущем счете равен 30 000 руб., а среднее квадратическое отклонение — 12 000 руб. У какого процента клиентов можно ожидать, что остатки на текущем счете будут более 50 000 руб.? Округлите процент до двух знаков после запятой.

```python
average_balance = 30000
standard_deviation = 12000
exp_balance = 50000
```

```python
# cdf(x, loc=0, scale=1): Кумулятивная функция распределения

proc_people = (1 - st.norm.cdf(x= exp_balance, loc=average_balance, scale=standard_deviation)) * 100
print('Процент клиентов:', proc_people.round(2))
```

```python
# sf(x, loc=0, scale=1): Функция выживания (также определяется как 1 - cdf, но sf иногда точнее)

c = st.norm.sf(x=50000, loc=30000, scale=12000)
print((c*100).round(2))
```

Посчитайте матрицу ковариаций признаков X из предыдущего упражнения. 

Из матрицы ковариаций посчитайте матрицу корреляций. 

Напишите свой класс, в котором будет две функции: расчет матрицы ковариаций и расчет матрицы корреляций. 

Используйте для этого только матричные операции. 

*Использовать напрямую методы библиотек — запрещено*.

```python
class Matrix():
    def covariance_matrices(x):
        cov_x = [[],[]] # Создаем пустую матрицу
        x0 = x[0].mean() # Считаем среднее первого вектора х
        x1 = x[1].mean() # Считаем среднее второго вектора х
        x_x0 = x[0] - x0 # Вычитаем среднее х из всего 1 ветора (получаем массив)
        x_x1 = x[1] - x1 # Вычитаем среднее х из всего 2 ветора (получаем массив)
        cov_x[0] = np.append(cov_x[0] ,np.sum(x_x0*x_x0)/(len(x_x0*x_x0)-1)) # Ковариация 1 ветора от 1 вектора
        cov_x[0] = np.append(cov_x[0] ,np.sum(x_x0*x_x1)/(len(x_x0*x_x1)-1)) # Ковариация 1 ветора от 2 вектора
        cov_x[1] = np.append(cov_x[1] ,np.sum(x_x1*x_x0)/(len(x_x1*x_x0)-1)) # Ковариация 2 ветора от 1 вектора
        cov_x[1] = np.append(cov_x[1] ,np.sum(x_x1*x_x1)/(len(x_x1*x_x1)-1)) # Ковариация 2 ветора от 2 вектора
        return cov_x # Возвращаем матрицу ковариаций
        
    def correlation_matrices(x):
        corr_x = [[],[]] # Создаем пустую матрицу
        x0 = x[0].mean() # Считаем среднее первого вектора х
        x1 = x[1].mean() # Считаем среднее второго вектора х
        x_x0 = x[0] - x0 # Вычитаем среднее х из всего 1 ветора (получаем массив)
        x_x1 = x[1] - x1 # Вычитаем среднее х из всего 2 ветора (получаем массив)
        corr_x[0] = np.append(corr_x[0] ,cov_x[0][0]/math.sqrt((np.sum(x_x0*x_x0)/(len(x_x0*x_x0)-1))*(np.sum(x_x0*x_x0)/(len(x_x0*x_x0)-1))))
        corr_x[0] = np.append(corr_x[0] ,cov_x[0][1]/math.sqrt((np.sum(x_x0*x_x0)/(len(x_x0*x_x0)-1))*(np.sum(x_x1*x_x1)/(len(x_x1*x_x1)-1))))
        corr_x[1] = np.append(corr_x[1] ,cov_x[1][0]/math.sqrt((np.sum(x_x1*x_x1)/(len(x_x1*x_x1)-1))*(np.sum(x_x0*x_x0)/(len(x_x0*x_x0)-1)))) 
        corr_x[1] = np.append(corr_x[1] ,cov_x[1][1]/math.sqrt((np.sum(x_x1*x_x1)/(len(x_x1*x_x1)-1))*(np.sum(x_x1*x_x1)/(len(x_x1*x_x1)-1))))
        return corr_x
        
```

```python
cov_x = Matrix.covariance_matrices(X) #Запускаем функцию расчета ковариации
print("Матрица ковариации: \n", np.array(cov_x),"\n")

corr_x = Matrix.correlation_matrices(X) #Запускаем функцию расчета корреляции
print("Матрица корреляции: \n", np.array(corr_x))
```

```python
example_picture = cv2.imread('/content/drive/MyDrive/School21/day09/svd.png')
picture_src = cv2.cvtColor(example_picture, cv2.COLOR_BGR2RGB)
print ("Вы должны будете получить что-то подобное:\n")
plt.imshow(picture_src)
plt.axis('off') #Отключаем отображение осей
plt.show()
```

```python
u, s, v = np.linalg.svd(picture_src_gray, full_matrices=False) # Разложение по сингулярным значениям
 
vector = [1, 2, 5, 10, 50, 100] # Задаем кол-во векторов для отображения
plt.figure(figsize=(30, 30), dpi=50) # Задаем размер отображаемых картинок
 
for i in range(len(vector)): # Создаем цикл для создания картинок с нужным кол-вом векторов
    low_rank = u[:, :vector[i]] @ np.diag(s[:vector[i]]) @ v[:vector[i], :] 
    #Собираем картинку из сингулярных значений с нужным кол-вом векторов
    plt.subplot(3, 2, i+1) # задаем ориентацию для картинок
    plt.imshow(low_rank, cmap='gray') #Устанавливаем карту цветов для отображения картинки в matplotlib
    
    plt.title(f'Cингулярный вектор из разложения SVD, первые = {vector[i]}') #Выводим подпись к картинке
   # plt.axis('off') #Отключаем отображение осей
```
