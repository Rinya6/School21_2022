3. Произведите агрегации с маркой и моделью:

- отобразите медианные штрафы в разрезе только марок (марка – медианный штраф внутри этой марки);
- отобразите медианные штрафы, сгруппированные по маркам и моделям (комбинация марка и модель – медианный штраф этой комбинации);
- отобразите количество штрафов, имеющихся в группировках по маркам и моделям, чтобы понять, можно ли доверять полученным медианным значениям;
- отобразите минимальные и максимальные штрафы, сгруппированные по маркам и моделям для того, чтобы лучше понять вариацию;
- отобразите стандартное отклонение штрафов, сгруппированных по маркам и моделям для того, чтобы лучше понять вариацию.

- [pandas.DataFrame.drop (by index)](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html)
- [pandas.DataFrame.groupby (index)](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)

```python
# отобразите медианные штрафы в разрезе только марок (марка – медианный штраф внутри этой марки)

df.drop(columns="Refund").groupby(by='Make', dropna=False).median(['Fines'])
```

```python
# отобразите медианные штрафы, сгруппированные по маркам и моделям (комбинация марка и модель – медианный штраф этой комбинации)

df.drop(columns="Refund").groupby(by=['Make', 'Model'], dropna=False).median()
```

```python
# отобразите количество штрафов, имеющихся в группировках по маркам и моделям, чтобы понять, можно ли доверять полученным медианным значениям

df.drop(columns="Refund").groupby(by=['Make', 'Model'], dropna=False).count()
```

```python
# отобразите минимальные и максимальные штрафы, сгруппированные по маркам и моделям для того, чтобы лучше понять вариацию;
```

- [pandas.core.groupby.DataFrameGroupBy.agg](https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html)

```python
import numpy as np
```

```python
df.drop(columns="Refund") \
  .groupby(by=['Make', 'Model'], dropna=False) \
  .agg(lambda x: (min(x), max(x), ['min', 'max']))
```

```python
# отобразите стандартное отклонение штрафов, сгруппированных по маркам и моделям для того, чтобы лучше понять вариацию.

df.drop(columns="Refund").groupby(by=['Make', 'Model'], dropna=False).std()
```

- [pandas.Index.value_counts](https://pandas.pydata.org/docs/reference/api/pandas.Index.value_counts.html)
- 

```python
# отобразите регистрационные номера автомобилей, 
# сгруппированные по количеству штрафов в порядке убывания (мы хотим найти тех, кто чаще всего нарушал правила дорожного движения)

df.index.value_counts()
```

```python
# выберите из исходного датафрейма все строки, 
# соответствующие регистрационному номеру с самым большим количеством нарушений (мы хотим проанализировать эти данные более подробно)

df_max = df[df.index.value_counts() == df.index.value_counts().max()]
df_max
```

```python
# отобразите регистрационные номера автомобилей, 
# сгруппированные по сумме штрафов в порядке убывания (мы хотим найти тех, кто заплатил самую большую сумму штрафов)

df_max_one = df.groupby('CarNumber').sum()\
                   .sort_values('Fines', ascending=False)
df_max_one
```

```python
# выберите из исходного датафрейма все строки, 
# соответствующие регистрационному номеру с самой большой суммой штрафов (мы хотим проанализировать эти данные более подробно)

df_one = df[df.index == df_max_one.head(1).index.values[0]]
df_one
```

```python
# вывести на экран таблицу, отвечающую на вопрос: 
# есть ли регистрационные номера, которые были установлены на разных моделях автомобилей?

df_group = df.drop(columns=['Refund', 'Fines']).groupby(['CarNumber', 'Make', 'Model'], dropna=False).count()
df_group
```

```python
df_group_list =list(df_group.index)
for i in range(1, len(df_group_list)):
    if df_group_list[i][0] == df_group_list[i-1][0]:
        print(df_group_list[i][0])
```
