```python
import pandas as pd
```

```python
# Загрузаем файл feed-views.log в датафрейм

df_one = pd.read_csv('/content/drive/MyDrive/School21/day10/task1/data/feed-views.log', sep = '\t')
```

```python
# изменить разделитель (на ",") и сохранить получившийся результат в другой файл (feed-views_new.log)

df_one.to_csv('/content/drive/MyDrive/School21/day10/task1/data/feed-views_new.log', index=False, sep = ',')
```

<h2>
1. Используя метод read_csv:

*   отфильтруйте строки с индексами 2 и 3 с помощью аргумента skiprows (мы заведомо знаем, что эти наблюдения были фиктивными);
*   отфильтруйте последние 2 строки таблицы с помощью аргумента skipfooter (мы знаем, что эти наблюдения тоже были фиктивными);
*   присвойте столбцам следующие имена: datetime, user;
*   используйте datetime в качестве столбца индекса.

```python
df_new = pd.read_csv('/content/drive/MyDrive/School21/day10/task1/data/feed-views_new.log', 
                     skiprows=[2,3],                 # фильтруем строки с индексами 2 и 3 с помощью аргумента skiprows
                     skipfooter=2, engine='python',  # фильтруем последние 2 строки таблицы с помощью аргумента skipfooter
                     names=['datetime', 'user'],     # присваиваем столбцам имена: datetime, user
                     index_col='datetime')           # используем datetime в качестве столбца индекса
df_new
```
