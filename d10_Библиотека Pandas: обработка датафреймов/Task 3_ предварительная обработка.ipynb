4. Работа с пропущенными значениями

- Проверьте, сколько значений отсутствует в каждом столбце.
- Удалите все столбцы с более чем 500 пропущенными значениями, используя аргумент thresh.
- Замените все пропущенные значения в столбце Refund предыдущим для данной ячейки значением в этом столбце, используйте аргумент method, проверьте, сколько значений теперь отсутствует в каждом столбце.
- Замените все пропущенные значения в столбце Fines средним значением этого столбца (исключите значения NA/null при вычислении среднего значения), проверьте теперь количество пропущенных значений в каждом столбце.

- [pandas.DataFrame.isna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isna.html)
- [pandas.DataFrame.dropna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html)
- [pandas.DataFrame.fillna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html)

```python
# проверьте, сколько значений отсутствует в каждом столбце.

df.isna().sum()
```

```python
# удалите все столбцы с более чем 500 пропущенными значениями, используя аргумент thresh.

df.dropna(axis=1, thresh=500, inplace=True)
df.isna().sum()
```

```python
# замените все пропущенные значения в столбце Refund предыдущим для данной ячейки значением в этом столбце, 
# используйте аргумент method, проверьте, сколько значений теперь отсутствует в каждом столбце.

df.Refund.fillna(method="ffill", inplace=True)
df.isna().sum()
```

```python
# замените все пропущенные значения в столбце Fines средним значением этого столбца 
# (исключите значения NA/null при вычислении среднего значения), проверьте теперь количество пропущенных значений в каждом столбце.

df.Fines.fillna(df.Fines.mean(), inplace=True)
df.isna().sum()
```

<h2>
5. Разбейте название марки и модели автомобиля на отдельные столбцы.

- Используйте метод apply как для разбиения, так и для извлечения значений в новые колонки Make и Model.
- Удалите столбец Make_n_model.
- Сохраните датафрейм в JSON-файл auto.json в следующем формате:
  * [{"CarNumber":"Y163O8161RUS","Refund":2.0,"Fines":3200.0,"Make":"Ford","Model":"Focus"},{"CarNumber":"E432XX77RUS","Refund":1.0,"Fines":6500.0,"Make":"Toyota","Model":"Camry"}]

- [pandas.DataFrame.apply](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)
- [pandas.DataFrame.to_json](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_json.html)

```python
# используйте метод apply как для разбиения, так и для извлечения значений в новые колонки Make и Model.

df['Make_n_model'] = df['Make_n_model'].apply(func=lambda x: x.split())
df
```

```python
df['Make'] = df['Make_n_model'].apply(lambda x: x[0])
df['Model'] = df['Make_n_model'].apply(lambda x: x[-1] if len(x) > 1 else None)
```

```python
df
```

```python
# удалите столбец Make_n_model

df.drop(columns='Make_n_model', inplace=True)
df
```

```python
# Сохраните датафрейм в JSON-файл auto.json в следующем формате:
# [{"CarNumber":"Y163O8161RUS","Refund":2.0,"Fines":3200.0,"Make":"Ford","Model":"Focus"},
#  {"CarNumber":"E432XX77RUS","Refund":1.0,"Fines":6500.0,"Make":"Toyota","Model":"Camry"}]

df.to_json('/content/drive/MyDrive/School21/day10/task3/data/auto.json', orient='records')
```
