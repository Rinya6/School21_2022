2. Итерирование: во всех следующих подзадачах необходимо рассчитать значение для каждой строки по формуле fines/refund*year, создать новый столбец с полученными данными и измерить время расчета, используя волшебную команду %%timeit в ячейке.

- Цикл: напишите функцию, которая выполнит перебор по датафрейму, используя for i in range(0, len(df)), iloc и append() в отношении списка, присвойте результат выполнения функции новому столбцу в датафрейме.
- Сделайте это с помощью функции iterrows().
- Сделайте это с помощью функции apply() и лямбда-функции.
- Сделайте это с помощью объектов Series из датафрейма.
- Сделайте это так же, как и в предыдущей подзадаче, но с помощью метода .values.

- [Что такое %timeit в Python?](https://stackoverflow.com/questions/29280470/what-is-timeit-in-python)
- [pandas.DataFrame.iterrows](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html)
- [pandas.DataFrame.apply](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)
- [pandas.Series.values](https://pandas.pydata.org/docs/reference/api/pandas.Series.values.html)

```python
# Цикл: напишите функцию, которая выполнит перебор по датафрейму, 
# используя for i in range(0, len(df)), iloc и append() в отношении списка, 
# присвойте результат выполнения функции новому столбцу в датафрейме.

def function_range():
  new_column = []
  for i in range(0, len(df)):
    new_column.append(df.iloc[i,2] / df.iloc[i,1] * df.iloc[i,5])
  return new_column
  
```

```python
%%timeit
df['New_column'] = function_range()
```

```python
df
```

```python
# Сделайте это с помощью функции iterrows()

%%timeit
df['New_column'] = [row[1]['Fines'] / row[1]['Refund'] * row[1]['Year'] for row in df.iterrows()]
```

```python
df
```

```python
# Сделайте это с помощью функции apply() и лямбда-функции

%%timeit
df['New_column'] = df.apply(lambda x: x[2] / x[1] * x[5], axis =1)
```

```python
df
```

```python
# Сделайте это с помощью объектов Series из датафрейма.

%%timeit
df['New_column']=df.Fines/df.Refund*df.Year
```

```python
df
```

```python
# Сделайте это так же, как и в предыдущей подзадаче, но с помощью метода .values

%%timeit
df['New_column']=df.Fines/df.Refund*df.Year.values
```

- [Типы данных](https://pyprog.pro/data_types.html)

- [pandas.to_numeric](https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html)
- [pandas.DataFrame.astype](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html)

```python
import numpy as np
```

```python
# выполните команду df.info(memory_usage='deep'), обратите внимание на Dtype и объем используемой оперативной памяти

df.info(memory_usage='deep')
```

```python
# скопируйте с помощью команды copy() исходный датафрейм в другой под названием optimized

optimized = df.copy()
```

```python
# выполните приведение типа вниз с float64 по float32 для всех столбцов

optimized = optimized.astype({'Fines': np.float32, 'New_column': np.float32})

#optimized.info(memory_usage='deep')
```

```python
# выполните приведение вниз с int64 до наименьшего возможного числового dtype

optimized.Refund = pd.to_numeric(optimized.Refund, downcast='integer')
optimized.Year = pd.to_numeric(optimized.Year, downcast='integer')

#optimized.info(memory_usage='deep')
```

```python
# запустите info(memory_usage='deep') теперь для нового датафрейма, 
# обратите внимание на Dtype и объем используемой оперативной памяти.

optimized.info(memory_usage='deep')
```
