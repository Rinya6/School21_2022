## 0. Импорты

````python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
````

## 1. Препроцессинг

1. Загрузите снова тот же датафрейм, как и в прошлом задании.


````python
df = pd.read_csv('/content/drive/MyDrive/School21/day14/data/day-of-week-not-scaled.csv')
df
````

````python
X = df.drop(columns = ['dayofweek'])
y = df.dayofweek
````

2. Воспользуйтесь `train_test_split` с параметрами `test_size=0.2`, `random_state=21` и получите `X_train`, `y_train`, `X_test`, `y_test`. Используйте дополнительный параметр `stratify`. 

````python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, 
                                                    random_state=21,
                                                    stratify=y)
````

1. Обучите SVM, используя параметры, которые оказались лучшими в прошлом задании.

gridsearch.best_params_:


 {'C': 10, 'class_weight': None, 'gamma': 'auto', 'kernel': 'rbf'}

````python
svm = SVC(C=10, 
          class_weight = None, 
          gamma = 'auto', 
          kernel = 'rbf', 
          random_state=21, 
          probability=True).fit(X_train, y_train)
````

````python
y_predict = svm.predict(X_test)
````

````python
print('accuracy is ', round(accuracy_score(y_test, y_predict),5))
print('precision is ', round(precision_score(y_test, y_predict, average='weighted'),5))
print('recall is ', round(recall_score(y_test, y_predict, average='weighted'),5))
print('roc_auc is ', round(roc_auc_score(y_test, svm.predict_proba(X_test), average='weighted', multi_class='ovo'),5))
````

## 6. Функция

1. Напишите функцию, которая принимает на вход список разных моделей и соответствующий модели список параметров в виде словаря. Функция возвращает словарь, который содержит все 4 метрики, расмотренных в этом задании, для каждой модели из списка. 
 {'model1': {'accuracy': 0.7, 'precision': 0.5, ... },'model2': {...}}

````python
model = {
                        SVC: {'C': 10,
                            'class_weight': None,
                            'gamma': 'auto', 
                            'kernel': 'rbf',
                            'random_state': 21, 
                            'probability': True},
    DecisionTreeClassifier: {'class_weight': 'balanced', 
                             'criterion': 'gini', 
                             'max_depth': 22,
                             'random_state': 21},
    RandomForestClassifier: {'class_weight': None,
                             'criterion': 'gini',
                             'max_depth': 28,
                             "n_estimators": 50,
                             'random_state': 21}
}
````

````python
def function (model):

  result = {}
  for i in model:
    parameters = model[i]
    models = i(**parameters)
    fit = models.fit(X_train, y_train)
    y_predict=models.predict(X_test)

    result[i] = {}
    
    result[i]['accuracy'] = accuracy_score(y_test, y_predict)
    result[i]['precision'] = precision_score(y_test, y_predict, average='weighted')
    result[i]['recall'] = recall_score(y_test, y_predict, average='weighted')
    result[i]['roc_auc'] = roc_auc_score(y_test, fit.predict_proba(X_test), average='weighted', multi_class='ovo')
    
    
  return result
 
````
