```python
pd.io.sql.read_sql("SELECT * from checker LIMIT 3", con)
```

```python
df = pd.io.sql.read_sql('''
              SELECT uid, DATE(timestamp) as date, count(timestamp) as count_date
              FROM checker
              WHERE status in ('ready') and
                    uid LIKE 'user_%' NOT LIKE 'admin_%' and
                    labname = 'project1'
              GROUP BY uid, date
              
              
''', con)
df
```

```python
df_count = df.copy()
df_count = df_count.sort_values(['date', 'uid'])\
                   .pivot(index='uid', columns='date', values='count_date')\
                   .fillna(0) \
                   .cumsum(axis=1) \
                   .transpose().sort_index().transpose()
df_count.reset_index(inplace=True)
df_count
```

```python
df_count.shape
```

```python
commit_max = df_count.drop(columns='uid').max().max()
commit_max 
```

```python
pio.templates.default = "simple_white"

numOfRows = df_count.shape[0] 
numOfCols = df_count.shape[1] 
numOfFrames = numOfCols - 1
xaxis_range = [0,numOfFrames + 2]

x_init = np.array([1])

initial_data = []
for user_ind in range(27):
    y_axis = np.array(df_count.iloc[user_ind,0])
    initial_data.append(go.Scatter(x =x_init, y = y_axis,mode = "lines",name = df_count.uid[user_ind]))
initial_max = commit_max

frames = []
for f in range(1,numOfFrames+1):
    x_axis = np.arange(1,f+1)
    curr_data = []
    title_names = []
    for user_ind in range(27):
        user = df_count.uid[user_ind]
        y_axis = np.array(df_count.iloc[user_ind,1:f+1])
        curr_data.append(go.Scatter(x = x_axis, y = y_axis,mode = "lines+markers", name = user))
    curr_frame = go.Frame(data = curr_data)
    frames.append(curr_frame)

 
figure = go.Figure(
   
    data = initial_data,
    layout = {
        "title":'Dynamic of commits per user in project1', 
        "xaxis":{"range":[2,20], "visible":True, "showline":False, "showgrid":True},
        "yaxis":{"range":[0,170], "visible":True, "showline":False, "showgrid":True},
        "updatemenus":[{"type":"buttons","buttons":[{"method":"animate","label":"play", "args":[None]}]}],
        "autosize":False,
        "width":1050,
        "height":700
             },
    frames = frames

    )

figure.update_layout(title={
                     'y':0.9,
                     'x':0.05,
                     'font': {'size':30}})

figure.update_layout(plot_bgcolor='lightgrey')
figure.show()
```
