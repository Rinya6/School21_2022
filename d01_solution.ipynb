```python
df.CLNT_JOB_POSITION_TYPE.unique()
```

```python
df.APP_POSITION_TYPE.unique()
```

```python
# сортировка

spisok = [x for x in list(df.CLNT_JOB_POSITION_TYPE.unique()) if x == x]
spisok.sort()
spisok
```

```python
spisok_с = [x for x in list(df.APP_POSITION_TYPE.unique()) if x == x]
spisok_с.sort()
spisok_с
```
