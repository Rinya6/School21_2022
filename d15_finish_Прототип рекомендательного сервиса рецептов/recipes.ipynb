```python
def get_content(html):
    soup = BeautifulSoup(html,'html.parser')
    items = soup.find('article', class_='recipe-content-card')
    cards = []
    
    for item in items:
       # if items.find('header', class_='summary').find('strong', class_='tag').get_text() == "recipefrom Gourmet":
        cards.append(
                    {
#                         'i': items.find('header', class_='summary').find('strong', class_='tag').get_text(),
                        'title': items.find('h4', class_='hed', itemprop='name').get_text(),
                        'rating': items.find('dl', class_='recipes-ratings-summary').get('data-reviews-rating'),
                        'link_recipe': HOST + items.find('h4', class_='hed').find('a').get('href')
                    }
                )
        
        return cards
    
```
